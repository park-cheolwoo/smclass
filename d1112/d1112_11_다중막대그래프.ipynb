```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np

matplotlib.rcParams["font.family"] = "Malgun Gothic"
matplotlib.rcParams["axes.unicode_minus"] = False
matplotlib.rcParams["font.size"] = 15
```

```python
df = pd.read_excel("score.xlsx")
```

```python
# 다중막대그래프 - 옆으로 여러개 막대그래프
df.shape
# 배열 [0,1,2,3,4,5,6,7]
index = np.array(df.index)
```

```python
w = 0.25
plt.bar(index,df['국어'],width=w,label='국어')
plt.bar(index-w, df["영어"], width=w, label="영어")
plt.bar(index-2*w, df["수학"], width=w, label="수학")
plt.xticks(index)
plt.legend(ncols=3)
plt.ylim(0,120)

for i in range(8):
    plt.text(index[i], df["국어"][i], df["국어"][i],ha='center')
    plt.text(index[i] - w, (df["영어"][i]), (df["영어"][i]), ha="center")
    plt.text(index[i]-2*w, (df["수학"][i]), (df["수학"][i]), ha="center")
plt.show()
```
