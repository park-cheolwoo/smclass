```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd

matplotlib.rcParams["font.family"] = "Malgun Gothic"
matplotlib.rcParams["axes.unicode_minus"] = False
matplotlib.rcParams["font.size"] = 15
```

```python
# x축 : 지원번호 y축 : 국어
# 선색상 : red, marker색상 : yellow, 테두리 색상 : blue, 선타입 : dash

df = pd.read_excel('score.xlsx')
```

```python
x = df.index.values
k = df['국어']
e = df['영어']
m = df['수학']
# plt.plot(x,y,color='red',ls="--",marker='o',mec='blue',mfc='yellow',ms='10')
plt.figure(figsize=(20,5),facecolor='yellow') # figsize : 그래프 화면 크기 조절, dpi : 그래프 해상도 설정(dpi가 클수록 이미지 크기 커짐) # facecolor : 배경색 설정
plt.plot(x,k,'ro--',label='국어',mec='blue',mfc='yellow',ms='10')
plt.plot(x,e,'bv-',label='영어',mec='blue',mfc='yellow',ms='10')
# plt.plot(x,m,'kx--',label='수학',mec='blue',mfc='yellow',ms='10')
plt.legend()
plt.show()
# plt.savefig('graph2.png',dpi=300) # 그래프 파일 저장
```
