```python
import pandas as pd
```

```python
# 상단 3개를 삭제후 가져오기
# index는 행정기관으로 지정해서 df로 변환해서 출력
# 위 5개만 출력

df = pd.read_excel("연령별인구현황01.xlsx", index_col="행정기관", skiprows=3,usecols='B,E:Y') # usecols : 컬럼을 선택해서 가져옴 'B,D,E', 'B,E:Y', 'A:E' (엑셀 상단 컬럼명과 동일)
df
```

```python
df['0~4세']
```

```python
# df[0] : 에러
# df : 전체를 보는 명령어
# df['컬럼명'] : 컬럼을 보는 명령어
# df[슬라이싱] : row를 보는 명령어 = df.iloc[번호] = df.loc[이름]
df.index
df[0:1]
# 데이터 replace함수를 적용, 타입을 변경 : astype
df.iloc[0] = df.iloc[0].str.replace(",","").astype(int) # 숫자 타입 >> describe(수학적 통계) 가능
df.describe()
```

```python
## 남자부부만 가져옴
## 여자 부분만 가져와서 출력
## 위 3개를 제외하고 행정기관 index 여자부분만 가져옴

df_w = pd.read_excel('연령별인구현황01.xlsx',index_col='행정기관',skiprows=3,usecols='B,AB:AV')
df_w
df_w['0~4세.1']
df_w.columns = df.columns # 여자컬럼과 남자컬럼명 다르기에 통일
df_w.columns

# 천단위,int타입 변경
df_w.iloc[0] = df_w.iloc[0].str.replace(",","").astype(int)
df_w.iloc[0]
```
