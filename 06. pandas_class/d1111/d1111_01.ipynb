```python
# 함수 : if문을 쓸 수 있음
## 50점 이하에 있는 학생의 점수를 +20점
filt = df['국어']<=50
df.loc[filt,'국어'] = df.loc[filt,'국어']+20
def func(value) :
  if value < 50:
    return (value + 20)
  else : 
    return value

df['국어'] = df['국어'].apply(func)
df
```

```python
# 1. 요일 컬럼 추가
# 2. 0-6까지의 숫자를 랜덤으로 입력하시오
# 3. 함수를 사용해서 0:일요일 1:월요일~6:일요일
import random
df['요일'] = 0
def func(value):
  value = random.randint(0,6)
  if value == 0: return '월요일'
  elif value == 1: return '화요일'
  elif value == 2: return '수요일'
  elif value == 3: return '목요일'
  elif value == 4: return '금요일'
  elif value == 5: return '토요일'
  elif value == 6: return '일요일'
df['요일'] = df['요일'].apply(func)
df
# df.iloc는 숫자만 가능
```
