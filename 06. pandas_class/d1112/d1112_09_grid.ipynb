```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd

matplotlib.rcParams["font.family"] = "Malgun Gothic"
matplotlib.rcParams["axes.unicode_minus"] = False
matplotlib.rcParams["font.size"] = 15
```

```python
df = pd.read_excel('score.xlsx')
df
```

```python
plt.plot(df['이름'],df['키'])
plt.xticks(rotation=45)
# 그래프 안내선 추가 axis : 축 방향 linestyle : 선 스타일
plt.grid(axis='both',linestyle='--',linewidth=1,color='purple',alpha=0.2)


# 그래프 값
for i in range(8):
    # plt.text(df.iloc[i, 1], df.iloc[i, 3]+0.7, df.iloc[i, 3])
    plt.text(df['이름'][i],df['키'][i]+0.7,df['키'][i],ha='center')
plt.show()
```
