```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np

matplotlib.rcParams["font.family"] = "Malgun Gothic"
matplotlib.rcParams["axes.unicode_minus"] = False
matplotlib.rcParams["font.size"] = 15
```

```python
df = pd.read_csv('drinks.csv')
# drinks.csv   total_servings    servings  
#  
# df.dropna(inplace=True)
df['total_servings'] = 0
df['total_servings'] = df['beer_servings']+df['spirit_servings']+df['wine_servings']

plt.figure(figsize=(10,5))
x = df['country']
y = df['total_servings']
plt.bar(x,y)
for i in df['country']:
  if i == 'South Korea':
    plt.bar(i,y,color='red')
plt.show()

plt.bar(x,y.sort_values())
for i in df["country"]:
    if i == "South Korea":
        plt.bar(i, df.iloc[i][6], color="red")
```
