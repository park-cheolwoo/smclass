```python
import pandas as pd
import numpy as np

df = pd.read_excel("score.xlsx", index_col="지원번호")
df

# 학교 구로고 > 가산고, 디지털고 > 단지고로 변경
# 4번 sw특기를 Django로 변경
# 5번 sw특기를 python 변경
# SW특기 소문자로 변경
# 합계컬럼 추가
# 결과컬럼 추가해서 270 이상은 Pass, 아니면 fail

df['학교'].rename({"구로고":"가산고","디지털고":"단지고"})
df.loc[['4번','5번'],'SW특기'] = ['Django','python']
df['SW특기'] = df['SW특기'].str.lower()
df['합계'] = 0
df['합계'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
df['결과'] = 'Fail'
df.loc[df['합계'] >=270,"결과"] = 'Pass'
df['학교']=df['학교']+'등학교'

# 문자열 = 문자열+문자열
df


```
