```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np

matplotlib.rcParams["font.family"] = "Malgun Gothic"
matplotlib.rcParams["axes.unicode_minus"] = False
matplotlib.rcParams["font.size"] = 15
```

```python
df = pd.read_excel("score.xlsx", index_col="지원번호")
```

```python
def cm_func(x):
  return str(x)+'cm'

df['키'] = df['키'].apply(cm_func)
```

```python
df['키']
```

```python
# 50점 아래에 있는 값은 50으로 리턴해서 출력
def k_func(x):
  if x<50: return 50
  return x


for i in range(8):
  print(k_func(df['국어'][i]))
```

```python
def func(x):
    return x/2

plt.figure()
plt.bar(df['이름'],df['국어'])
plt.bar(df['이름'],df['영어'],bottom=df['국어'])
plt.bar(df['이름'],df['수학'],bottom=df['국어']+df['영어'])
for i in range(8):
    plt.text(df["이름"][i], func(df["국어"][i])-2.3, df["국어"][i],ha='center')
    plt.text(df["이름"][i], func(df["영어"][i]+df['국어'][i]+df['국어'][i]), df["영어"][i],ha='center')
    plt.text(
        df["이름"][i],
        func(
            df["수학"][i]
            + df["영어"][i]
            + df["국어"][i]
            + df["영어"][i]
            + df["국어"][i]
        ),
        df["수학"][i],
        ha="center",
    )
plt.show()
```
