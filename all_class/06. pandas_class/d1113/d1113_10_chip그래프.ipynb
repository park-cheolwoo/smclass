```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np

matplotlib.rcParams["font.family"] = "Malgun Gothic"
matplotlib.rcParams["axes.unicode_minus"] = False
matplotlib.rcParams["font.size"] = 15

df = pd.read_csv("chipotle.tsv", sep="\t")
df
```

```python
# 메뉴별 주문개수를 10개만 뽑아서 막대그래프로 출력
data = df.groupby('item_name')['order_id'].count().head(10)
# 평균치 추가
data['mean'] = int(df.groupby('item_name')['order_id'].count().mean())

x = data.index
y = data
mean = df.groupby("item_name")["order_id"].count().mean()

plt.figure(figsize=(30,5))
graph = plt.bar(x,y,label = '주문 횟수',width=0.2)
graph[len(data)-1].set_color('red')
plt.plot(data.index,[mean]*len(data),color='k',ls='--')
plt.xticks(rotation=90)
for i in range(11):
  plt.text(x[i],y[i]+2,y[i],ha='center')
```
