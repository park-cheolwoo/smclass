```python
import pandas as pd
import numpy as np
```

```python
# 컬럼 2개 구성 - 가중치1, 가중치2
# 전체 가중치1 컬럼 (사망+부상)*1.1
# 2014년 사망,부상 가중치2 (사망+부상)*1.2
# 2015년 사망,부상 가중치2 (사망+부상)*1.5
# 2015년 사망,부상 가중치2 (사망+부상)*1.3
df1 = pd.read_csv('2014년졸음운전교통사고.csv',encoding='euc-kr')
df2 = pd.read_csv('2015년졸음운전교통사고.csv',encoding='euc-kr')
df3 = pd.read_csv('2016년졸음운전교통사고.csv',encoding='euc-kr')
df = pd.concat([df1,df2,df3])
df
df['가중치1'] = 0
df['가중치1'] = df['사망(명)']+df['부상(명)']
df['가중치2'] = 0
# 조건식 2014년에 해당되는 조건식 구성
filt1 = df['구분'].str[0:4] == '2014'
filt2 = df['구분'].str[0:4] == '2015'
filt3 = df['구분'].str[0:4] == '2016'
# 2014에 해당되는 가중치2 계산해서 데이터 수정
df.loc[filt1,'가중치2'] = (df.loc[filt1,'사망(명)']+df.loc[filt1,'부상(명)'])*1.2
df.loc[filt2,'가중치2'] = (df.loc[filt2,'사망(명)']+df.loc[filt2,'부상(명)'])*1.5
df.loc[filt3,'가중치2'] = (df.loc[filt3,'사망(명)']+df.loc[filt3,'부상(명)'])*1.3
# df.loc[filt1,'가중치2'] = df['사망(명)']+df['부상(명)']*1.2
# df.loc[filt2,'가중치2'] = df['사망(명)']+df['부상(명)']*1.5
# df.loc[filt3,'가중치2'] = df['사망(명)']+df['부상(명)']*1.3
```
