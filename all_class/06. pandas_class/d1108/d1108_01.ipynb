```python
#  row : df.loc / df.iloc
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
# loc = index값으로 검색
# iloc = index번호로 검색
df.loc['1번'] # 1개의 row를 가져옴
df.loc[['1번','5번']] # 2개의 row를 가져옴
df.loc['1번','국어'] # 1개 row의 1개 컬럼만 가져옴
df.loc[['1번','5번'],['국어','영어']] # 1번,5번 row의 국어,영어 성적만 가져옴
# 슬라이싱
df.loc['1번':'5번','국어':'사회'] 

df.iloc[0]
df.iloc[[0,2],[3,4]] # 1번,2번 학생의 국어,영어 성적을 출력
# 슬라이싱
df.iloc[1:4] # 2번,3번,4번 학생 출력
df.iloc[1:4, 0:3]  # 2번,3번,4번 학생의 이름,학교,키 출력
```

```python
# 데이터 조건 검색
filt = df['키']>=185 # 조건식 생성
filt
df[filt] # 조건식 대입 columns
df[~filt]  # not에 대한 결과값 출력 
```

```python
df.loc[~filt] # rows
df.loc[filt,['수학','국어']] # rows
```

```python
# and 연산자 &, or 연산자 |
filt = (df['국어']>=70) & (df['키']<190)
df[filt]
```

```python
# 문자열 검색(startswith,contains,replace,lower,upper)
filt = df['이름'].str.startswith('김')
df[filt]

# contains : 해당값이 포함되어 있는 단어 검색
filt = df['이름'].str.contains('수')
df.loc[filt,['이름','학교','키']]

# isin() 포함되어 있는지 확인
langs = ['Python','Java']
filt = df['SW특기'].isin(langs)
df[filt]

# lower() : 해당 값을 모두 소문자로 변환
# isin() : 괄호 안에는 리스트 타입만 가능
langs = ['python','java']
filt = df['SW특기'].str.lower().isin(langs)
df[filt]

# 포함이 되어 있으면 모두 검색
filt = df['SW특기'].str.contains('Java',na=False) # NaN 값을 처리하지 않으면 에러
df[filt]

# 소문자로 변경 후 j가 포함되어 있는 정보 검색

```

```python
# dropna
# axis : index/column how : any/all
# axis : 축방향으로 삭제 index(row방향) columns(column 방향)
# how : any(1개라도 있으면 모두 row/column 삭제), all(모두 NaN이어야 row/column 삭제)
df.dropna() # 기본 default : index,any
df.dropna(axis='index',how='all') # 모두 NaN row가 없어서 삭제가 안됨
df.dropna(axis='columns',how='all') # 학교 컬럼이 모두 NaN이므로 삭제됨
df.dropna(axis='index',how='any') # 학교 컬럼이 모두 NaN이어서 모든 row가 삭제됨
df.dropna(axis='columns',how='any') # 학교,SW특기 컬럼에 NaN이 있기에 2개 컬럼 삭제
```
