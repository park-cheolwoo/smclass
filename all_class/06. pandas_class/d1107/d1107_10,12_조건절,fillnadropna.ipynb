```python
import pandas as pd
import numpy as np



df = pd.read_excel("score.xlsx", index_col="지원번호")

df
```

```python
# 조건
filt = df['키'] >=185
df[filt]
df[~filt] # 조건에 해당되지 않는 값만 출력
df[df['키']>=185]
```

```python
df[df["키"] >= 185] # 키가 185 이상 모든 컬럼 출력
df.loc[df['키']>=185] # 키가 185 이상 모든 컬럼 출력, 컬럼을 지정할 수 있음
df.loc[df['키']>=185,"이름"] # 이름 컬럼만 출력
df.loc[df['키']>=185,["이름","키","SW특기"]] # 이름, 키, SW특기 출력
df['키']>=185 # True,False의 값으로 출력
```

```python
# 조건식 and & or |
filt = (df['키'] >=185) & (df['학교'] == "가산고")
df[filt]
```

```python
# or 조건식
df.loc[(df['키']<170) | (df['키']>200)]
```

```python
## 문자열(str) 조건 -> '김'으로 시작되는 : startwith
# 문자열 함수 : endwith,replace,strip
filt = df["이름"].str.startswith("강")
df[filt]
df[df['이름'].str.startswith("김")]


## '수' 가 포함되어 있는 이름 검색 : contains
filt = df['이름'].str.contains("수")
df[filt]

df['SW특기'].str.lower() # 소문자 변환
df['SW특기'].str.upper() # 대문자 변환
df['SW특기'].str # 대문자 변환

df['SW특기']
langs = ['Python','Java']
# isin = 리스트에 있는 단어가 포함되어 있는지 체크
filt = df['SW특기'].isin(langs)
df[filt] # 대소문자를 구분함

# 소문자 치환
langs = ['python','java']
filt = df['SW특기'].str.lower().isin(langs)
df[filt]

```

```python
## Java라는 글자 있는지 체크
## na = True : NaN True 인식, na=False : NaN False 인식
filt = df["SW특기"].str.contains("Java",na=False) # null값이 포함되어있으면 에러
df[filt] 
```

```python
# 결측치(NaN 데이터)
# NaN 데이터 채우기 : fillna()
df.fillna(' ',inplace=True)
df

# 학교컬럼에 NaN 데이터를 입력 : np.nan

df.fillna('없음') # inplace=True, 모든 NaN 값에 해당데이터를 입력
df['SW특기'].fillna('없음',inplace=True) # 해당 컬럼에만 적용
df
```

```python
df = pd.read_excel("score.xlsx", index_col="지원번호")
# df["학교"] = np.nan
df
```

```python
# NaN 데이터 삭제 : dropna()
df.dropna()
```

```python
# axis 축 : indeex 방향으로 삭제
df.dropna(axis='index',how='any') # NaN이 들어있는 4,5번 row 데이터가 삭제
```

```python
# dropna : NaN 데이터 삭제 , axis : 축 - index(row),columns(column)
# how default : any - NaN값이 1개라도 있으면 삭제 , all - 전부 NaN일 때 삭제
df["학교"] = np.nan
df.dropna(axis = 'columns',how='all')
```

```python
# fillna : NaN 데이터 해당값으로 채워줌
# dropna : NaN 데이터를 삭제
# - axis : 축 index(row방향으로 삭제)/columns(column방향으로 삭제)
# - how : 어떤 경우에 삭제할지 결정 any(1개라도 NaN 값이 있으면 삭제) / all(모두  NaN값일 경우 삭제) 
```
