```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np

matplotlib.rcParams["font.family"] = "Malgun Gothic"
matplotlib.rcParams["axes.unicode_minus"] = False
matplotlib.rcParams["font.size"] = 15
```

```python
df = pd.read_excel("score.xlsx", index_col="지원번호")
```

```python
# 원그래프
values = [30,25,20,13,10,2] # 100 퍼센트를 기준으로 자동 계산
explode = [0.3,0.1,0,0,0,0]
labels=['python','java','javascript','oracle','html','etc']
colors = ['plum','beige','skyblue','bisque','khaki','peachpuff']
plt.title('프로그램 사용률')
# pie : 원형그래프 labels : 글자출력 explode : 원형그래프에서 분리 counterclock : 시계반대방향 (False : 시계방향)
# startangle : 시작되는 각도 위치 (90 = 12시부터 시작 / 180 = 9시부터 시작) autopct : 데이터값 출력(%.1f : 데이터 수치의 포맷 설정, %% : 문자'%' 출력)
# color : 색상 선택
plt.pie(values,labels=labels,explode=explode,counterclock=False,startangle=90, autopct='%.1f %%', colors=colors)
plt.legend(loc=(1.2,0.3),title='언어별 선호도') # 범례가 그래프를 모두 가리므로 절대값으로 설정
# plt.legend(loc='upper right')
plt.show()
```
