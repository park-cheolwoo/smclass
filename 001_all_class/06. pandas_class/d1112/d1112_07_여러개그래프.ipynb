```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd

matplotlib.rcParams["font.family"] = "Malgun Gothic"
matplotlib.rcParams["axes.unicode_minus"] = False
matplotlib.rcParams["font.size"] = 15
```

```python
x = [1, 2, 3]
y = [2, 4, 8] # sm매출액
y2=[5,1,3] # LG매출액
y3=[1,2,5] # SK매출액

plt.figure(figsize=(10,5))
plt.plot(x,y,label='SM매출액',marker='o',ls="-.")
plt.plot(x,y2,label='LG매출액',marker='x',ls="--")
plt.plot(x,y3,label='SK매출액',marker='v',ls=":")
for i in range (3):  
  plt.text(x[i],y[i]+0.2,y[i],ha='center')
  plt.text(x[i],y2[i]+0.2,y2[i],ha='center')
  plt.text(x[i],y3[i]+0.2,y3[i],ha='center')
# for idx,y2 in enumerate(y2):
#   plt.text(x[idx],y2+0.2,y2)
# for idx,y3 in enumerate(y3):
#   plt.text(x[idx],y3+0.2,y3)
plt.title("기업별 매출금액")
plt.legend(ncol=3) # 기본 : 세로출력, ncol=3
plt.show()
```
