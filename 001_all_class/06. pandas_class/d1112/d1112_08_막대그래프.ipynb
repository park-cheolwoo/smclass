```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd

matplotlib.rcParams["font.family"] = "Malgun Gothic"
matplotlib.rcParams["axes.unicode_minus"] = False
matplotlib.rcParams["font.size"] = 15
```

```python
label = ['홍길동','유관순','강감찬']
values = [180,187,184]
colors = ['red','green','blue']

plt.figure(figsize=(5,3))
plt.title('막대그래프')
# plt.plot(label,values) # 선그래프
# plt.bar(label,values,color='red') # 막대그래프 # color : 색상(1개)
plt.bar(label,values,color=colors,width=0.1,label='학생 키') # 막대그래프 # color : 색상(여러개) # width : 바 크기 # alpha : 투명도 조절
plt.xticks(label,['1번','2번','3번'],rotation=45) # 라벨 변경, 라벨 기울기 설정
plt.yticks([0,50,100,150,200,250])
plt.legend(loc="upper center")
plt.show()
```
