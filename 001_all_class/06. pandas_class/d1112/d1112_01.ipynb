```python
import pandas as pd
import numpy as np

df = pd.read_excel('score.xlsx',index_col = '지원번호')
df
```

```python
## 랜덤으로 0~6사이 랜덤숫자 입력
import random
df['요일']=0
```

```python
# 특정 셀 수정
df.loc['1번','요일']=5
df.loc['2번','요일']=1

df.iloc[0, 9] = 5
df.iloc[1, 9] = 5

# 컬럼 전체를 수정
df.loc['요일'] = 5
```

```python
df.loc['1번','요일'] = random.randint(0,6)
df.iloc[0,9] = random.randint(0,6)
# 랜덤으로 0부터 6까지 숫자 입력
for i in range(8):
  df.iloc[i,9] = random.randint(0,6)
```

```python
def func(x):
  if x == 0: return '일요일'
  elif x == 1: return '월요일'
  elif x == 2: return '화요일'
  elif x == 3: return '수요일'
  elif x == 4: return '목요일'
  elif x == 5: return '금요일'
  else: return '토요일'
```

```python
df['요일']=df['요일'].apply(func)
```

```python
# 결과 컬럼 생성
df['결과'] = '불합격'

## 국어 점수가 60점 이상이면 합격으로 수정
filt = df['국어']>=60
df.loc[filt,'결과'] = '합격'
```

```python
# '키'+'cm' astype() 타입변경 후 문자열+문자열
df['키'] = df['키'].astype(str)+'cm'
df
```

```python
df = pd.read_csv("drinks.csv", sep=",")
df
```

```python
# 모든 술의 총량을 출력하시오
## total_servings / 전체 알코올 도수를 확인
df['total_servings'] = df['beer_servings']+df['spirit_servings']+df['wine_servings']
```

```python
df
```

```python
# 대륙별 total_servings를 구하시오
df.groupby('continent')['total_servings'].sum()
```

```python
filt = df['continent'].isnull()

df.loc[df['continent'].isnull(),'total_servings']
```

```python
## continent : 대륙 NaN값을 OT로 변경
df['continent'].fillna('OT',inplace=True)
```

```python
## 술 소비량 대비 알코올 비율이 어느 정도 되는지 출력
# 알코올 비율을 계산에서 컬럼을 추가
# 높은 나라별 20개를 출력
df['alchohol_rate'] = df['total_servings']/df['total_litres_of_pure_alcohol']
df.sort_values('alchohol_rate',ascending=False).head(20)
df.groupby('continent')['alchohol_rate'].sum().sort_values(ascending=False)
```

```python
## kor 포함되어 있는 나라 출력
filt = df['country'].str.lower().str.contains('kor',na=False)
df[filt][['country','alchohol_rate']]
```
