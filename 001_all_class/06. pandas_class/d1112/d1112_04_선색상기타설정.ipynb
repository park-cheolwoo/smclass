```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams["font.family"] = "Malgun Gothic"
matplotlib.rcParams["axes.unicode_minus"] = False
matplotlib.rcParams["font.size"] = 15
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
yy = [1,8,3]
plt.plot(
    x,
    y,
    label="인원 증가",
    color = "pink",
    linewidth=5,
    marker="o",
    markersize=10,
    markeredgecolor="red",
    alpha=0.3,
    mfc='green'
)  # color : 선색상, linewidth : 그래프 두께, markeredgecolor : 마커 테두리 선 색상(mec), markerfacecolor(mfc) : 마커 내용 색상
# plt.plot(x, yy,label='비용 증가',marker='+',markersize=10,ls="-") # marker : 지점 표시 o/x/v/+/*/d markersize : 마커 크기, linestyle(ls) : :/--/-./-
plt.plot(x,yy,'kv:',label='비용변동') # b:선색상 v:마커 ::점선
plt.title("라인그래프 Line", loc="center", size=50)  
plt.xlabel("X축 날짜", color="#e83420", loc="center")  
plt.ylabel("Y축 인원", color="blue", loc="center")

# 범례 출력 : 그래프 선 설명문구, loc: 범례 위치 upper/center/lower left/center/right
# plt.legend(loc='lower right')
plt.legend(loc=(0.2,0.5)) # 좌표값으로 위치 설정
```
