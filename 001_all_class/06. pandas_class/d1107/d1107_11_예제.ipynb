```python
# students.xlsx 파일 불러와서 평균 80점 이상, 국어 90점 이상 출력
```

```python
import pandas as pd
```

```python
df = pd.read_csv('students.csv',index_col='no')
df
```

```python
filt = (df['avg']>=80) & (df['kor']>=90)
df[filt]
```

```python
## 평균이 60점이상 또는 국어점수가 70점 이상
filt = (df['avg']>=60) | (df['kor']>=70)
df[filt]
```

```python
# 이름에 t가 들어가면서 평균점수가 80점 이상인 사람 출력
filt = (df['avg']>=80) & (df['name'].str.contains('t'))
df[filt]
df[filt][['name','kor']]
```
