```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np

matplotlib.rcParams["font.family"] = "Malgun Gothic"
matplotlib.rcParams["axes.unicode_minus"] = False
matplotlib.rcParams["font.size"] = 15

df = pd.read_excel('연령별인구현황_202410.xlsx',skiprows=3,usecols="B,E:Y",nrows=1,index_col='행정기관')
df
```

```python
# 연령별 인구피라미드 그리기
x = df.columns
# df.index
# df.rename({"전국  ":"전국"},inplace=True)
# df.loc['전국'] = df.loc['전국'].str.replace(",","").astype(int)/1000000
y = df.loc['전국'].values

plt.figure(figsize=(5,10))
plt.title('2024년 인구현황')
plt.xlabel('인구수(백만)')
plt.ylabel('연령대')
plt.barh(x,y,label='인구 수', color='skyblue')
plt.xlim(0,5)
for i in range(len(x)):
    plt.text(y[i] + 0.6, x[i], f"{y[i]*1000000:.0f}", ha="center", va="center")
```
