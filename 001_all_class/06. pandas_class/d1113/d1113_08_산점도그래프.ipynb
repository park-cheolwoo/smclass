```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np

matplotlib.rcParams["font.family"] = "Malgun Gothic"
matplotlib.rcParams["axes.unicode_minus"] = False
matplotlib.rcParams["font.size"] = 15

```

```python
df = pd.read_excel('score.xlsx')
df["학년"] = [3, 3, 2, 1, 1, 3, 2, 2]
```

```python
import random
```

```python
sizes = df['학년']*500
# 산점도그래프
plt.title('산점도 그래프')
plt.xlabel('영어')
plt.ylabel('수학')
# s : 산점도 크기 변경, c : 산점도 표시 컬럼, cmap: 색상표 사용
plt.scatter(
    df["영어"], df["수학"], s=sizes, c=df["학년"], cmap="plasma", alpha=0.7
)
#colorbar : 색상바 생성, 산점도 색상의 학년 표시 ticks : 범위 표시
plt.colorbar(ticks=[1,2,3],label='학년',shrink=0.5,orientation = 'horizontal')
```
