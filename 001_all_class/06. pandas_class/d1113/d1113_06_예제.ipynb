```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np

matplotlib.rcParams["font.family"] = "Malgun Gothic"
matplotlib.rcParams["axes.unicode_minus"] = False
matplotlib.rcParams["font.size"] = 15
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
# 학교별 인원을 원그래프로 출력
values = df.groupby('학교')['이름'].count()
labels = values.index
explode = [0.1,0]
wedgeprops = {'width':0.6,"edgecolor":"white","linewidth":5}
plt.pie(values,labels = labels,autopct='%.1f %%',startangle=90,wedgeprops=wedgeprops,pctdistance=0.7)
plt.show()
```
