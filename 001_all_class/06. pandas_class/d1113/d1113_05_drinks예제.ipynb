```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np

matplotlib.rcParams["font.family"] = "Malgun Gothic"
matplotlib.rcParams["axes.unicode_minus"] = False
matplotlib.rcParams["font.size"] = 15
```

```python
df = pd.read_csv('drinks.csv')
# total_servings 컬럼을 추가해서 대륙별 total_servings 원 그래프로 출력
df['total_servings'] = 0
df["total_servings"] = df["beer_servings"] + df["spirit_servings"] + df["wine_servings"]
values = df.groupby('continent')['total_servings'].sum().sort_values()
labels = df.groupby("continent")["total_servings"].sum().index
wedgegrops= {'width':0.6,"edgecolor":'w',"linewidth":2}
plt.pie(values, autopct = '%.1f %%',startangle=90,labels=labels, wedgeprops=wedgegrops,pctdistance=0.7)
plt.show()
```

```python
## 대륙별 total_litres_of_pure_alcohol 막대그래프로 출력하시오.
## 마지막에는 평균 막대그래프를 출력하시오.
df = pd.read_csv("drinks.csv")
# 1. 전체 total_litres_of_pure_alcohol 평균을 구하기
data = df["total_litres_of_pure_alcohol"].mean()
n_data = [data]*6
# 2. 대륙별 total_litres_of_pure_alcohol 평균을 구하기
# 'AF', 'AS', 'EU', 'OC', 'OT', 'SA'
con_df = df.groupby("continent")["total_litres_of_pure_alcohol"].mean()
con_df
# 3. 컬럼에 대륙별 평균을 추가하기 mean
con_df["mean"] = data
plt.plot(con_df.index,n_data,ls="--",color='black')
plt.grid(axis='y',ls='--',alpha=0.5)
# 4. 막대그래프를 출력
con_bar = plt.bar(con_df.index, con_df)
con_bar[len(con_df)-1].set_color('red')
```
