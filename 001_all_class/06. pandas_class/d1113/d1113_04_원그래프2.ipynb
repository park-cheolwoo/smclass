```python
values = [30, 25, 20, 13, 10, 2]  # 100 퍼센트를 기준으로 자동 계산
# explode = [0.3, 0.1, 0, 0, 0, 0]
wedgeprops = {'width':0.6,"edgecolor":"white","linewidth":2}
labels = ["python", "java", "javascript", "oracle", "html", "etc"]
colors = ["plum", "beige", "skyblue", "bisque", "khaki", "peachpuff"]
# wedgeprops : 가운데 구멍 생성(도넛 모형) / 숫자가 작을수록 구멍이 커짐
# edgecolor : 경계선 색상, linewidth : 경계선 두께
# pctdistance : 글씨를 바깥쪽으로 이동시킴

## 데이터 값이 2% 이하는 적용하지 않도록 함수 적용
# pct : values의 퍼센트 값 생성
def custom_autopct(pct):
    if pct >= 10: return f"{pct:.1f} %"
    # 문자 포맷팅 함수 : %
    else: return ""


plt.pie(values,labels=labels,colors=colors,autopct=custom_autopct,wedgeprops=wedgeprops,\
         startangle=90,counterclock=False,pctdistance=0.7)
plt.show()
```
