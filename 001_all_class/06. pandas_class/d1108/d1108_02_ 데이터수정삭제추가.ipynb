```python
import pandas as pd
import numpy as np

df = pd.read_excel("score.xlsx", index_col="지원번호")
df
```

```python
# 데이터 수정
## replace : 데이터값을 변경, inplace=True 지정해야 데이터가 저장됨
# df['학교'].replace({"구로고":"가산고"},inplace=True)
# df
```

```python
# df["학교"].replace({"구로고": "가산고","디지털고":"단지고"}, inplace=True)
# df

# 특정 컬럼값 변경
# 컬럼 = 변경값을 입력
df["SW특기"] = df["SW특기"].str.lower()
df['SW특기'].str.upper()
df['SW특기'].str.capitalize() # 첫글자 대문자
df["키"] = 200
df
```

```python
# 특정셀 값 변경
df.loc['1번','국어'] =100
df.loc["2번", ["국어", "영어"]] = 10
```

```python
df.info()
```

```python
# 컬럼명 = 문자+문자
df['학교'] = df['학교']+'등학교'
df
```

```python
df.loc[['2번','3번'],['국어','영어']] = 50
```

```python
# 데이터 추가
## 컬럼에 없는 컬럼 이름을 입력하면 컬럼이 추가됨
df['합계'] = 0
df['합계'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
df['평균'] = df['합계']/5
df
```

```python
# 결과 컬럼 추가
df['결과'] = 'Fail'
# 결과컬럼에 조건식을 사용해서 데이터 입력
filt = df['합계']>=400
df.loc[filt,'결과'] = 'Pass'
df
```

```python
df['합산'] = 100
## 국어점수가 90점 이상이면 df['합산']+100
filt = df['국어']>=90
df.loc[filt,'합산'] = df['합산']+100
df
```

```python
# 데이터 삭제 - drop
df.drop(columns=['합산'],inplace=True) # inplace=True
# 컬럼 2개 삭제
df.drop(columns = df.columns[3:5])
df.drop(columns=["합계", "평균", "결과"])
```

```python
# drop - index 삭제
df.drop(index='1번') # inplace=True
df.head(2)
```
