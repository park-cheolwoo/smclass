```python
import pandas as pd
import numpy as np

df = pd.read_csv("chipotle.tsv", sep="\t")
df
```

```python
## 메뉴별 주문수량을 출력하시오
df.groupby('item_name')['quantity'].count()
```

```python
## 메뉴별 총주문량을 출력하시오
df.groupby("item_name")["quantity"].sum()
```

```python
## 주문당 평균 계산금액을 출력하시오
# df["item_price"] = df["item_price"].str[1:].astype(float)
df['total'] = df['item_price']*df['quantity']
df.groupby("order_id")['total'].sum().mean()

# df.groupby("order_id")["item_price"].mean()
```

```python
# item별 가격 금액을 출력하시오
df.groupby('item_name')['item_price'].mean()
```

```python
## 주문별, 높은 가격 순으로 주문 개수 출력
df.groupby('order_id')[['total']].sum().sort_values('total',ascending=False)
```
