```python
import pandas as pd
import numpy as np

df = pd.read_excel("score.xlsx", index_col="지원번호")
```

```python
def add_cm(height):
    return str(height) + "cm"
```

```python
df['키'] = df['키'].apply(add_cm)
df
```

```python
df['SW특기'] = df['SW특기'].str.lower()
```

```python
def change(lang):
    if pd.notnull(lang):
        # pd.notnull(list) : list에서 null인지 아닌지 유무를 True/False로 확인
        # lang.str.upper() -> list에서 1개 str씩
        return lang.upper()
    else:
        return lang
```

```python
# 함수 호출 : apply
df['SW특기'] = df['SW특기'].apply(change)

df
```
